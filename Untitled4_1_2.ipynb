```python
# prompt: What does the scatter diagram developed indicate about the relationship between the two variables?

import numpy as np
import matplotlib.pyplot as plt

x=[20, 20, 30, 30, 40, 40, 50, 50]
y=[23, 22, 19, 15, 14, 17, 14, 12]
plt.plot(x, y)

# Calculate the correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Add the correlation coefficient to the plot
plt.title(f"Scatter Plot with Correlation: {correlation:.2f}")

plt.show()
```

```python
# prompt: Use the least squares method to develop the estimated regression equation

import matplotlib.pyplot as plt
import numpy as np
# Calculate the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the numerator and denominator of the slope
numerator = 0
denominator = 0
for i in range(len(x)):
  numerator += (x[i] - x_mean) * (y[i] - y_mean)
  denominator += (x[i] - x_mean) ** 2

# Calculate the slope and intercept
slope = numerator / denominator
intercept = y_mean - slope * x_mean

# Calculate the estimated values of y
y_estimated = []
for i in range(len(x)):
  y_estimated.append(slope * x[i] + intercept)

# Plot the data and the estimated regression line
plt.plot(x, y, 'o')
plt.plot(x, y_estimated)

# Add the equation of the estimated regression line to the plot
plt.title(f"Scatter Plot with Regression Line: y = {slope:.2f}x + {intercept:.2f}")

plt.show()
```
