```python
# prompt: Develop the estimated regression equation by computing the values of
# b0
#  and
# b1
#  using
# b1 =
# Σ(xi − x)(yi − y)
# Σ(xi − x)2
#  and
# b0 = y − b1x

import numpy as np

x = [1, 2, 3, 4, 5]
y = [3, 8, 5, 10, 14]

# Calculate the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the numerator and denominator of the slope formula
numerator = 0
denominator = 0
for i in range(len(x)):
  numerator += (x[i] - x_mean) * (y[i] - y_mean)
  denominator += (x[i] - x_mean) ** 2

# Calculate the slope and intercept
b1 = numerator / denominator
b0 = y_mean - b1 * x_mean

# Print the estimated regression equation
print(f"Estimated regression equation: y = {b0:.3f} + {b1:.3f}x")
```
